```python
import plotly.graph_objects as go

def make_graph(stock_data, revenue_data, company_name):
    # Create the stock price figure
    stock_fig = go.Figure()
    stock_fig.add_trace(go.Scatter(
        x=stock_data["Date"],
        y=stock_data["Close"],
        mode='lines',
        name=f"{company_name} Stock Price"
    ))
    stock_fig.update_layout(
        title=f"{company_name} Stock Prices Over Time",
        xaxis_title="Date",
        yaxis_title="Stock Price"
    )

    # Create the revenue figure
    revenue_fig = go.Figure()
    revenue_fig.add_trace(go.Bar(
        x=revenue_data["Date"],
        y=revenue_data["Revenue"],
        name=f"{company_name} Revenue"
    ))
    revenue_fig.update_layout(
        title=f"{company_name} Revenue Over Time",
        xaxis_title="Date",
        yaxis_title="Revenue"
    )
    

    return stock_fig, revenue_fig
```
